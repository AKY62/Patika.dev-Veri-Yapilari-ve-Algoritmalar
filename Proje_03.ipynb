# Proje 3

[7, 5, 1, 8, 3, 6, 0, 9, 4, 2] dizisinin Binary-Search-Tree aşamaları aşağıdaki gibidir:

ilk adım, 7'yi root alarak sırasıyla diğer sayıları yerleştirmeye başıyoruz:

                    7
5, 7'den küçük olduğundan soluna yerleşecek:

                    7
                   /
                  5

1'i yerleştiriyoruz (5'ten küçük, o yüzden 5'in soluna):

                    7
                   /
                  5
                 /
                1




8'i yerleştiriyoruz (7'den büyük, o yüzden 7'nin sağına):

                    7
                   / \
                  5   8
                 /
                1

3'ü yerleştiriyoruz (1'den büyük, o yüzden 1'nin sağına):

                    7
                   / \
                  5   8
                 /
                1
                 \
                  3
                
6'yı yerleştiriyoruz (5'ten büyük, o yüzden 5'nin sağına):

                    7
                   / \
                  5   8
                 / \
                1   6
                 \
                  3

0'ı yerleştiriyoruz (1'den küçük, o yüzden 1'in soluna):

                    7
                   / \
                  5   8
                 / \
                1   6
               / \
              0   3

                    7
                   / \
                  5   8
                 / \   \
                1   6   9
               / \
              0   3

                    7
                   / \
                  5   8
                 / \   \
                1   6   9
               / \
              0   3
                   \
                    4
